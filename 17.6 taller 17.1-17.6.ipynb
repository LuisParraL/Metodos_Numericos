```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
xi=[2.0, 4.0, 6.0, 7.0, 10.0, 11.0, 14.0, 17.0, 20.0]
yi=[1.0, 2.0, 5.0, 2.0, 8.0, 7.0, 6.0, 9.0, 12.0]

n=len(xi)
mult, sum1,sumax,sumay,t,z,ly,lx=0,0,0,0,[],[],[],[]

for i in range (n):
    sumax+=xi[i]
    sumay+=yi[i]
    mult+=xi[i]*yi[i]
    sum1+=xi[i]**2

expo=sumax**2
a1=(n*mult-(sumax*sumay))/(n*sum1-expo)
xmedia=sumax/n
ymedia=sumay/n
a0=ymedia-(a1*xmedia)
#print a0, a1
for j in range (n+10):
    y=a0+a1*j
    t.append(j)
    z.append(y)


for k in range (1,n):
    #lny=math.log(a1)+math.log(k)+1
    lny=math.log(z[k])
    ly.append(lny)
    lx.append(math.log(k))
              
#print ly
plt.plot(t,z, label='recta y=a0+a1*x')
plt.plot(xi,yi, 'ro', label='xi,yi')
plt.plot(lx,ly, label='ln(y) / ln(x)')
plt.axis([0,18,0,10])
plt.xlabel('(x)')
plt.ylabel('f(x)')
plt.legend(loc='upper left', numpoints=1)
plt.grid()
plt.show()
```
