```python
import math
import matplotlib.pyplot as plt
from tabulate import tabulate
t,z,fc1,fc2,xl,xu,vv,tab,fc11,fc22,xu1,xl1,tab1=[],[],1.0,1.0,0.5,1.0,0.5794,[],1.0,1.0,1.0,0.5,[]

Es=0.1
Es1=0.002
print "                              Bisecci贸n                                  "
print "~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~"
for i in range (60):
    fx=0.7*math.pow(i,5)-8*math.pow(i,4)+44*math.pow(i,3)-90*math.pow(i,2)+82*math.pow(i,1)-25
    t.append(i)
    z.append(fx)
    
def funcion (num):
    fx=0.7*math.pow(num,5)-8*math.pow(num,4)+44*math.pow(num,3)-90*math.pow(num,2)+82*math.pow(num,1)-25
    return fx
i=0
xr=0
while (fc1*fc2)!=0:
    xra=xr
    xr=(xu+xl)/2
    fc1=funcion(xl)
    fc2=funcion(xr)
    if ((fc1*fc2)<0):
        xu=xr
    if ((fc1*fc2)>0):
        xl=xr
    i+=1
    Et=vv-xr
    Erpv=math.fabs((Et/vv)*100)
    Ea=math.fabs(((xr-xra)/(xr))*100)
    tab.append([i,xr,Erpv,Ea])
    if (Ea<Es):
        break
print tabulate (tab,headers=["Iteracion", "Raiz", "Error R. Porc. Verd (%)", "Error Aprox.(%)"],tablefmt='orgtbl')
print " "
print "                        Falsa Posici贸n                                  "
print "~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~"
def funcion2 (num1):
    fx1=0.7*math.pow(num1,5)-8*math.pow(num1,4)+44*math.pow(num1,3)-90*math.pow(num1,2)+82*math.pow(num1,1)-25
    return fx1

j=0
xr1=0
while (fc11*fc22)!=0:
    xra1=xr1
    f2=funcion2(xu1)
    f1=funcion2(xl1)
    xr1=xu1-(f2*(xl1-xu1))/(f1-f2)
    fc11=funcion2(xl1)
    fc22=funcion2(xr1)
    if ((fc11*fc22)<0):
        xu1=xr1
    if((fc11*fc22)>0):
        xl1=xr1
    j+=1
    Et1=vv-xr1
    Erpv1=math.fabs((Et1/vv)*100)
    Ea1=math.fabs(((xr1-xra1)/xr1)*100)
    tab1.append([j,xr1,Erpv1,Ea1])
    if (Ea1<Es1):
        break
print tabulate (tab1,headers=["Iteracion", "Raiz", "Error R. Porc. Verd (%)", "Error Aprox.(%)"],tablefmt='orgtbl')
plt.plot(t,z)
plt.plot(vv,0, 'cs', label='Raiz Exacta=0.5794')
plt.legend(loc = 'upper right', numpoints = 1)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
