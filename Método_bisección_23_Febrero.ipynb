```python
import math
import matplotlib.pyplot as plt
from tabulate import tabulate
m=68.1
g=9.8
v=40.0
t=10.0
t=[]
f=[]
tab=[]
xl=12.0
xu=16.0
xr=0.0
fc1=1.0
fc2=1.0
i=1
def funcion2(num):
    fc=((g*m)/num)*(1-math.exp(-(num/68.1)*10.0))-v
    return fc
    tab.append([fc])
while ((fc1*fc2)!=0):
    xr=(xl+xu)/2
    fc1=funcion2(xl)
    fc2=funcion2(xr)
    if ((fc1*fc2)<0):
        xu=xr
    if((fc1*fc2)>0):
        xl=xr
    c=i*1
    fc=((g*m)/c)*(1-math.exp(-(c/68.1)*10.0))-v
    f.append(fc)
    t.append(c)
    i+=1
    tab.append([c,xl,xu,(fc1*fc2),xr,fc])
print tabulate(tab,headers=["Iteracion", "xl", "xu", "F(xl)*F(xu)", "Raiz", "F(c)"],tablefmt='orgtbl')
plt.plot(t,f)
plt.xlabel('c')
plt.ylabel('f(c)')
plt.grid(True)
plt.show()
```
