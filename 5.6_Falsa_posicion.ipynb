```python
import math
from tabulate import tabulate
fc1,fc2,xu,xl,tab=1.0,1.0,1.3,0.0,[]

def funcion(num):
    fx=math.pow(num,10)-1
    return fx
i=0
xr=0
while ((fc1*fc2)!=0):
    xra=xr
    f2=funcion(xu)
    f1=funcion(xl)
    xr=xu-(f2*(xl-xu))/(f1-f2)
    Ea=math.fabs(((xr-xra)/xr)*100)
    fc1=funcion(xl)
    fc2=funcion(xr)
    if ((fc1*fc2)<0):
        xu=xr
    if((fc1*fc2)>0):
        xl=xr
    i+=1
    if (i==65):
        break
    tab.append([i,xr,xl,xu,fc1*fc2,Ea])
print tabulate (tab,headers=["Iteracion", "Raiz","xl","xu", "Condicion", "Error Aprox."],tablefmt='orgtbl')
```
