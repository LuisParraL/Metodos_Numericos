```python
import matplotlib.pyplot as plt
import math
%matplotlib inline
t=[]
log=[]
vec=[]
x,x0, x1,fx0,fx1=2.0,1.0,6.0,0.0, math.log(6)
x1b,fx1b=4.0, math.log(4)
x2,fx2,b0=6.0, math.log(6),fx0


f12a= fx0 + ((fx1-fx0)/(x1-x0))*(x-x0)
f12b = fx0 + ((fx1b-fx0)/(x1b-x0))*(x-x0)
b1=((fx1b-fx0)/(x1b-x0))
b2=((((fx2-fx1b)/(x2-x1b))-(b1))/(x2-x0))

f22=(b0+(b1*(x-x0))+b2*(x-x0)*(x-x1b))


ar1=[fx0,f12a,fx1]
ar2=[fx0,f12b,fx1b]
ar3=[fx0,f22,fx1b]

t1=[1,2,6]
t2=[1,2,4]
print "Estimación con interpolación lineal x1 = 6",f12a
print "Estimación con interpolación lineal x1 = 4",f12b
print "Estimación con interpolación cuadrática x1 = 4",f22
for i in range (1,8):
    vec.append((b0+(b1*(i-x0))+b2*(i-x0)*(i-x1b)))
    t.append(i)
    log.append(math.log(i))
    
plt.title('Interpolacion lineal y cuadratica')
plt.plot(t,log, marker='o', linewidth=2.0,linestyle='--',label='f(x)= ln(2)')
plt.plot(t1,ar1, marker='*',linewidth=2.0 , linestyle='-.',label='Estimacion x1=6')
plt.plot(t2,ar2,marker='v',linewidth=2.0 , linestyle='-.',label='Estimacion x1=4')
plt.plot(t,vec, marker='s',linewidth=2.0 , linestyle='-.')
plt.plot(2,f22, 'p', label='f2(2)=0.5658')
plt.axis([1,7,0,2])
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend(loc='upper left', numpoints=1)
plt.grid(True)
plt.show()
```

```python
#interpolación cuadrática para ln (x), x>=1
import math
import matplotlib.pyplot as plt
%matplotlib inline
t,log, vec=[],[],[]
x=input('ingrese el punto donde se va a hacer la interpolacion (x>=1) ')
x0=input('Ingrese punto x0 -> ')
x1=input('ingrese punto x1 -> ')
x2=input('ingrese punto x2 -> ')
if (x>0 and x0>0 and x1>0 and x2>0):
    fx0=math.log(x0)
    fx1=math.log(x1)
    fx2=math.log(x2)
    b0=fx0
    b1=((fx1-fx0)/(x1-x0))
    b2=((((fx2-fx1)/(x2-x1))-(b1))/(x2-x0))

    f12a= fx0 + ((fx1-fx0)/(x1-x0))*(x-x0)

    f22=(b0+(b1*(x-x0))+b2*(x-x0)*(x-x1))


    ar1=[fx0,f12a,fx1]

    t1=[x0,x,x1]

    for i in range (1,11):
        vec.append((b0+(b1*(i-x0))+b2*(i-x0)*(i-x1)))
        t.append(i)
        log.append(math.log(i))
    plt.plot(t,log, label='f(x)= ln(x)')
    plt.plot(t,vec, label='Estimacion cuadratica')
    plt.plot(t1,ar1, marker='*',linewidth=2.0 , linestyle='-.',label='Estimacion lineal')
    plt.plot(x,f22, 'p', label='aproximacion')
    plt.legend(loc='upper left', numpoints=1)
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.grid(True)
    plt.show()
    print "Error relativo porcentual verdadero igual a : ", ((math.log(x)-f22)/(math.log(x)))*100,"%"
else:
    print "Error, ingrese número mayores a 0 "
```
