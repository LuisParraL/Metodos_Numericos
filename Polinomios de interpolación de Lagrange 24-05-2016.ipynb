```python
import math
import matplotlib.pyplot as plt
import matplotlib.pyplot as plt2
%matplotlib inline
n,xi,x,y,t,y1,y2= 3,2,[],[],[],[],[]
x = [1,4,6]

for k in range(n):
    y.append(math.log(x[k]))

def lagrange(x,y,n,xi):
    suma=0.0
    for i in range(n):
        product = y[i]
        for j in range(n):
            if (i!=j):
                product = product*(xi-x[j])/(x[i]-x[j])
        suma = suma+product 
    lagrange = suma
    return lagrange

print "ln(2) por interpolacion de lagrange de primer orden ",lagrange(x,y,2,xi)
print "ln(2) por interpolacion de lagrange de segundo orden ",lagrange(x,y,3,xi)

i = 0.0
while(i<50):
    i += 0.5
    y1.append(lagrange(x,y,2,i))
    y2.append(lagrange(x,y,3,i))
    t.append(i)
    
plt.title('ln(2) por interpolacion de Lagrange de primer orden')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.plot(t,y1, label="Aprox. por interp de Lagrange 1er orden ln(2)")
plt.plot(xi,math.log(xi), 'ro',linewidth=3.0, label="Valor Verdadero="+str(math.log(xi)))
plt.legend(loc="upper right", numpoints=1)
plt.axis([-5, 50, -5, 40])
plt.grid(True)
plt.show()
plt2.title('ln(2) por interpolacion de Lagrange de segundo orden')
plt2.ylabel('f(x)')
plt2.xlabel('x')
plt2.plot(t,y2, label="Aprox. por interp de Lagrange 2do orden ln(2)")
plt2.plot(xi,math.log(xi), 'ro',linewidth=3.0, label="Valor Verdadero="+str(math.log(xi)))
plt2.axis([0, 20, -10, 10])
plt2.legend(loc="upper right", numpoints=1)
plt2.grid(True)
plt2.show()
```
